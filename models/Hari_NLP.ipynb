```python
import speech_recognition as sr
from googletrans import Translator
from gtts import gTTS
from playsound import playsound

def recognize_speech():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)
    try:
        print("Recognizing...")
        text = recognizer.recognize_google(audio, language='en')
        print("Recognized text:", text)
        return text
    except sr.UnknownValueError:
        print("Could not understand audio.")
        return ""
    except sr.RequestError as e:
        print("Error with the service; {0}".format(e))
        return ""

def translate_to_tamil(text):
    translator = Translator()
    translation = translator.translate(text, src='en', dest='ta')
    return translation.text

def translate_to_english(text):
    translator = Translator()
    translation = translator.translate(text, src='ta', dest='en')
    return translation.text

def speak_text(text, lang):
    tts = gTTS(text=text, lang=lang)
    tts.save("output.mp3")
    playsound("output.mp3")

def main():
    translate_to_tamil_flag = True  
    while True:
        print("Speak something: ")
        if translate_to_tamil_flag:
            original_text = recognize_speech()
            if original_text.lower() == 'q':
                break
            if original_text:
                translated_text = translate_to_tamil(original_text)
                print("Tamil translation:", translated_text)
                speak_text(translated_text, 'ta')
                translate_to_tamil_flag = False
        else:
            original_text = recognize_speech()
            if original_text.lower() == 'q':
                break
            if original_text:
                translated_text = translate_to_english(original_text)
                print("English translation:", translated_text)
                speak_text(translated_text, 'en')
                translate_to_tamil_flag = True

if __name__ == "__main__":
    main()
```
